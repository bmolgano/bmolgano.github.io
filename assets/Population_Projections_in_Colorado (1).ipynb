```python
import matplotlib.pyplot as plt
import seaborn as sns

# Quick check for null values
print(df.isnull().sum())
```

```python
df = pd.read_csv('Population_Projections_in_Colorado_20250713.csv')
```

```python
df.columns
```

Total Projected Population Over Time

```python
plt.figure(figsize=(12,6))
df_grouped = df.groupby('year')['totalPopulation'].sum().reset_index()

sns.lineplot(x='year', y='totalPopulation', data=df_grouped)
plt.title("Total Projected Population in Colorado Over Time")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```

Top 10 Counties by Population in a Selected Year (Bar Plot)

```python
selected_year = 2050
df_year = df[df['year'] == selected_year].sort_values('totalPopulation', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='totalPopulation', y='county', data=df_year, palette='Blues_r')
plt.title(f"Top 10 Counties by Total Population in {selected_year}")
plt.xlabel("Total Population")
plt.ylabel("County")
plt.show()
```

Population Growth for One County Over Time

```python
county_name = 'Denver County'
df_county = df[df['county'] == county_name]

plt.figure(figsize=(12,6))
sns.lineplot(x='year', y='totalPopulation', data=df_county)
plt.title(f"Projected Population Growth for {county_name}")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```

Heatmap: Population Projections by County and Year

```python
df_pivot = df.pivot_table(index='county', columns='year', values='totalPopulation', aggfunc='sum')

plt.figure(figsize=(14,10))
sns.heatmap(df_pivot, cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap of Population Projections by County and Year")
plt.xlabel("Year")
plt.ylabel("County")
plt.show()
```

Histogram of Total Population Values (All Counties, All Years)

```python
plt.figure(figsize=(10,6))
sns.histplot(df['totalPopulation'], bins=30, kde=True)
plt.title("Distribution of County Total Population Values (All Years)")
plt.xlabel("Total Population")
plt.ylabel("Frequency")
plt.show()
```

Boxplot: Total Population by Year

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='year', y='totalPopulation', data=df)
plt.xticks(rotation=45)
plt.title("Total Population Projections Distribution by Year")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```
